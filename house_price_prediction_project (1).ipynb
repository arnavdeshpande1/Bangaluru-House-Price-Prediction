```python
import pandas as pd
import numpy as np

df = pd.read_csv('Bengaluru_House_Data.csv')
df.head()
```

```python
df1 = df.drop(['area_type','availability','society','bath'],axis='columns')
df1.head()
```

```python
df1.isnull().sum()
```

```python
df2 = df1.dropna()
df2.isnull().sum()
```

```python
df2['bhk'] = df2['size'].apply(lambda x: x.split(' ')[0])
df2.head()
```

```python
df2['bhk'].unique()
```

```python
df2['total_sqft'].head(30)
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None

  
```

```python
df3 = df2.copy()
```

```python
df3.total_sqft = df3.total_sqft.apply(convert_sqft_to_num)
```

```python
df3.head()
```

```python
df3.loc[30]
```

```python
df4 = df3.copy()
df4['price_per_sqft'] = df4['price']*100000/df4['total_sqft']
```

```python
df4.head()
```

```python
len(df4['location'].unique())
```

```python
df4.location = df4.location.apply(lambda x: x.strip())

location_statistics = df4.groupby('location')['location'].agg('count')
location_statistics
```

```python
location_statistics = df4.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_statistics
```

```python
location_statistics_less_than_10 = location_statistics[location_statistics<=10]
location_statistics_less_than_10
```

```python
df4.location = df4.location.apply(lambda x: 'other' if x in location_statistics_less_than_10 else x)
len(df4.location.unique())
```

```python
df5 = df3.copy()
```

```python
df5.head()
```

```python
df5['price_per_sqft'] = df5['price']*100000/df5['total_sqft']
```

```python
df5.head()
```

```python
df5.groupby('location')['location'].agg('count')
location_stats =  df5.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
location_stats[location_stats<=10]
```

```python
location_stats_less_than_10 = location_stats[location_stats<=10]
location_stats_less_than_10
```

```python
len(df5.location.unique())
```

```python
df5.location = df5.location.apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
len(df5.location.unique())
```

```python
df6 = df5.copy()
```

```python
df6.head()
```

```python
df6.price_per_sqft.describe()
df6.shape
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df7 = remove_pps_outliers(df6)
df7.shape
df7.head()
```

```python
df8=df7.copy()
df8 = df8.drop(['size','price_per_sqft'],axis='columns')
df8 = df8.astype({"bhk": 'float64'})
df8.head(3)
```

```python
dummies = pd.get_dummies(df8.location)
dummies.head(3)
```

```python
df9 = pd.concat([df8,dummies.drop('other',axis='columns')],axis='columns')
df9.head()
```

```python
df10 = df9.drop('location',axis='columns')
df10.head(5)
```

```python
df10.shape
```

```python
X = df10.drop(['price'],axis='columns')
X.head(3)
```

```python
y = df10.price
y.head(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python

from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)

```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X, y, cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,y)
```


Based on above results we can say that LinearRegression gives the best score. Hence we will use that.

```python
def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return lr_clf.predict([x])[0]
```
